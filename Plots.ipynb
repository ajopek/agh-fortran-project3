```python
import pandas as pan
import matplotlib.pyplot as plot
```

```python
# Load data
df = pan.read_table('result.data', sep=r"\s+", header=0)
```

```python
df.loc[(df['func'] == "mm")].sort_values(by='n').groupby(['n', 'type']).agg({'time': 'mean'}).unstack().plot()
```

```python
df.loc[(df['func'] == "gauss")].sort_values(by='n').groupby(['n', 'type']).agg({'time': 'mean'}).unstack().plot()
```
